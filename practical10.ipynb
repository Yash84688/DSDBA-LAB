```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

iris = sns.load_dataset("iris")

print("Dataset Information:")
print(iris.dtypes)

print("\nDataset Summary:")
print(iris.describe())
```

```python
iris.hist(figsize=(10, 6), bins=20, edgecolor='black')
plt.suptitle("Feature Distributions (Histogram)", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris)
plt.title("Box Plot for Each Feature")
plt.show()
```

```python
Q1 = iris.iloc[:, :-1].quantile(0.25)
Q3 = iris.iloc[:, :-1].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = iris[((iris.iloc[:, :-1] < lower_bound) | (iris.iloc[:, :-1] > upper_bound)).any(axis=1)]
print("\nOutliers Detected:")
print(outliers)
```
