```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('tips')
```

```python
df.head()
```

```python
# heat map 
corr=df.corr()
sns.heatmap(corr)
```

```python
#heatmap with values
sns.heatmap(corr,annot=True)
```

```python
#joint plot for tip and total_bill
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
#jointplot hex
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
#joint plot for tip and total_bill with regression line
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#pairplot
sns.pairplot(df,hue='sex')
```

```python
#distplot
sns.distplot(df['tip'])
```

```python
#distplot
sns.distplot(df['tip'],kde='False')
```

```python
#rugplot
sns.rugplot(df['tip']) 
```

```python
#categorical plot
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
#countplot
sns.countplot(x='sex', data=df)
```

```python
#boxplot
sns.boxplot(x='sex', y='tip', data=df) 
```

```python
df.head(1)
```

```python
#box plot with hue
sns.boxplot(x='sex', y='tip', data=df, hue="smoker")
```

```python
# violin plot
sns.violinplot(x='total_bill', y='day', data=df)
```

```python
# violin plot with hue
sns.violinplot(x='total_bill', y='day', data=df,hue="smoker")
```
